```python
id = []
json_data = []
lastservices = []
for indice, fila in data.iterrows():
  opcional = False
  id.append(fila['id'])
  service = {
      'id': fila['id'],
      'number': fila['number'],
      'service_name': fila['service_name'],
      'services' : {}
  }

  if fila['AB'] == 1:
    service['services']['AB'] = {"AB":True}
    opcional = True

  if fila['AC'] == 1:
    service['services']['AC'] = {"AC":True}
    opcional = True

  if fila['AM'] == 1:
    service['services']['AM'] = {"AM":True}
    opcional = True

  if fila['AS'] == 1:
    service['services']['AS'] =  {"AS": True}
    opcional = True

  if fila['CT'] == 1:
    service['services']['CT'] =  {"CT": True}
    opcional = True

  if fila['EN'] == 1:
    service['services']['EN'] =  {"EN": True}
    opcional = True

  if fila['GA'] == 1:
    service['services']['GA'] =  {"GA": True}
    opcional = True

  if fila['IN'] == 1:
    service['services']['IN'] =  {"IN": True}
    opcional = True

  if fila['TE'] == 1:
    service['services']['TE'] =  {"TE": True}
    opcional = True

  if fila['TF'] == 1:
    service['services']['TF'] =  {"TF": True}
    opcional = True

  if fila['TM'] == 1:
    service['services']['TM'] =  {"TM": True}
    opcional = True

  if fila['TV'] == 1:
    service['services']['TV'] =  {"TV": True}
    opcional = True

  if fila['IN'] == 1 and fila['TE'] == 1 :
    if fila['service_name']  == 'INTERNET':
        service['services']['IN'] = {"IN": True, "cdServic": True}
        service['services']['TE'] = {"TE": True}
    if fila['service_name']  == 'TELEFONIA MOVIL':
        service['services']['TE'] = {"TE": True, "cdServic": True}
        service['services']['IN'] = {"IN" :True}
    opcional = True

  if fila['TE'] == 1 and fila['TV'] == 1 :
    if fila['service_name']  == 'TELEVISION':
        service['services']['TV'] = {"TV": True, "cdServic": True}
        service['services']['TE'] = {"TE": True}
    if fila['service_name']  == 'TELEFONIA MOVIL':
        service['services']['TE'] = {"TE": True, "cdServic": True}
        service['services']['TV'] = {"TV" :True}
    opcional = True

  if opcional:
    json_data.append(service)

  lastservices.append(service['services'])

structure = {
        'id' : id,
        #'info_account':json_data,
        'services':lastservices
}
```
